```python
import pymysql.cursors
import pandas as pd
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
host = 'den1.mysql6.gear.host'
db = 'situation'
usr = 'situation'
pwd = input("Enter your database password: ")

try: 
    connection = pymysql.connect(host=host, user=usr, password=pwd, database=db, 
                                 cursorclass=pymysql.cursors.DictCursor)
    print("Succesfully Connected!")
except pymysql.Error as err:
    print("Error %d: %s" % (err.args[0], err.args[1]))
    connection = None
```

```python
if connection:
    sql = "SHOW TABLES;"
    try:
        with connection.cursor() as cursor:
            cursor.execute(sql)
            tables = cursor.fetchall()
            print("Tablolar:")
            for table in tables:
                print(table)
    except pymysql.Error as err:
        print("Hata %d: %s" % (err.args[0], err.args[1]))
```

```python
# Access to necessary tables
table_names = [
    'c3_devicecalibration',
    'c3_device_properties',
    'c3_devices',
    'c3_devices_depr',
    'c3_devices_properties_temp'
]

data_frames = {}

if connection:
    for table_name in table_names:
        sql = f"SELECT * FROM {table_name};"
        try:
            data_frames[table_name] = pd.read_sql(sql, connection)
            print(f"{table_name} tablosundan veriler başarıyla çekildi!")
            print(data_frames[table_name].head())
        except pymysql.Error as err:
            print(f"Hata %d: %s - {table_name}" % (err.args[0], err.args[1]))

# Close the connection
connection.close()
```
