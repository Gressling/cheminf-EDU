```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA


df = data

# Summary statistics
print("Summary Statistics:")
print(df.describe())
print("\n\nMaxima:")
print(df.max())
print("\n\nMinima:")
print(df.min())
print("\n\nMean:")
print(df.mean())
#print("\n\nStddev:")
#print(df.std())


# Pairplot to visualize the relationships
#Here's a brief explanation of what sns.pairplot(df) does in more detail:2
#Scatter Plots: Off-diagonal plots show scatter plots between pairs of features. This helps visualize the relationship (e.g., linear, non-linear) between each pair of variables.
#Histograms/Density Plots: The diagonal plots show the distribution of each variable using histograms or density plots.
#Visual Patterns: It helps in identifying patterns, correlations, and potential outliers within the dataset.

sns.pairplot(df)
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# fix some weird decimal.Decimal representation of numbers to just use float
df = df.apply(pd.to_numeric, downcast='float')
# Create a 3D scatter plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
scatter = ax.scatter(df['temperature'], df['oxygen_flow_rate'], df['time'], c=df['final_carbon_content'], cmap='viridis')

# Labels and title
ax.set_xlabel('Temperature')
ax.set_ylabel('Oxygen Flow Rate')
ax.set_zlabel('Time')
ax.set_title('3D Scatter Plot')

# Color bar
cbar = fig.colorbar(scatter, ax=ax, pad=0.1)
cbar.set_label('Final Carbon Content')

# Show plot
plt.show()





# Principal Component Analysis (PCA)
pca = PCA(n_components=2)
components = pca.fit_transform(df[['temperature', 'oxygen_flow_rate', 'initial_carbon_content', 'time']])

pca_df = pd.DataFrame(data=components, columns=['PC1', 'PC2'])
pca_df['final_carbon_content'] = df['final_carbon_content']

print("\nPCA Explained Variance Ratio:")
print(pca.explained_variance_ratio_)

# Plot PCA
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='final_carbon_content', palette='viridis', s=100)
plt.title('PCA - Final Carbon Content')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
# plt.colorbar(label='Final Carbon Content')
plt.show()
```
